모델 성능 향상 방법 - K겹 교차 검증

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score
```

```python
import pandas as pd
```

```python
!git clone https://github.com/taehojo/data.git
```

```python
df = pd.read_csv('./data/sonar3.csv', header=None)
```

```python
X = df.iloc[:,0:60]
y = df.iloc[:,60]
```

```python
k = 5
kflod = KFold(n_splits=k, shuffle=True)
```

```python
# acc_score 리스트 초기화 (각 fold의 정확도 저장용)
acc_score = []
def model_fn():
  model = Sequential()
  model.add(Dense(24, input_dim=60, activation='relu'))
  model.add(Dense(10, activation='relu'))
  model.add(Dense(1, activation='sigmoid'))
  return model
```

```python
# k겹 교차 검증 시작
for train_index, test_index in kflod.split(X):
  # 훈련용 데이터와 검증용 데이터 분리
  X_train, X_test = X.iloc[train_index,:], X.iloc[test_index,:]
  y_train, y_test = y.iloc[train_index], y.iloc[test_index]

  # 새 모델 생성 (기존 학습 내용 초기화)
  model = model_fn()
  # 모델 컴파일 (이진 분류용 설정)
  model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
  # 모델 학습 수행
  history = model.fit(X_train, y_train, epochs=200, batch_size=10, verbose=0)
  # 검증용 데이터로 모델 성능 평가 (정확도 추출)
  accuracy = model.evaluate(X_test, y_test)[1]
  # 정확도 저장
  acc_score.append(accuracy)
```

```python
acc_score
```

```python
mean_acc = sum(acc_score) / 5
mean_acc
```

```python
df = pd.read_csv('./data/wine.csv', header=None)
```

```python
X = df.iloc[:, 0:12]
y = df.iloc[:, 12]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True)
```

```python
model = Sequential()
model.add(Dense(24, input_dim=12, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
history = model.fit(X_train, y_train, epochs=50, batch_size=500, validation_split=0.25)
```

```python
score = model.evaluate(X_test, y_test)
```

```python
print(score[1])
```

```python
hist_df = pd.DataFrame(history.history)
hist_df
```
