```python
from re import X
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from sklearn.metrics import accuracy_score
```

```python
train_data = pd.read_csv('https://storage.googleapis.com/harrykim-homepage/student_train_data.csv')
test_data = pd.read_csv('https://storage.googleapis.com/harrykim-homepage/student_test_data.csv')
```

```python
# head 구조 파악
train_data.head(5)
```

```python
# 각 열별 결측치 개수 확인

print(train_data[['weekly_hours', 'quiz_score_avg', 'discussion_posts', 'activity_score', 'login_frequency']].isnull().sum())
print(test_data[['weekly_hours', 'quiz_score_avg', 'discussion_posts', 'activity_score', 'login_frequency']].isnull().sum())
```

```python
# train weekly_hours 열에 결측치가 있는 행 제거
train_data = train_data.dropna(subset=['weekly_hours'])
print(train_data.isnull().sum())
```

```python
# test weekly_hours 열에 결측치가 있는 행 제거
test_data = test_data.dropna(subset=['weekly_hours'])
print(test_data.isnull().sum())
```

```python
# train test split
X_train = train_data.drop(columns=['name','gender','region','successful'], axis=1)
y_train = train_data['successful']
X_test = test_data.drop(columns=['name','gender','region', 'successful'], axis=1)
y_test = test_data['successful']
```

```python
model = Sequential()

model.add(Dense(128, input_dim=5, activation='relu', name='Dense1'))
model.add(BatchNormalization(name='BN1'))
model.add(Dropout(0.3, name='Dropout1'))

model.add(Dense(64, activation='relu', name='Dense2'))
model.add(BatchNormalization(name='BN2'))
model.add(Dropout(0.3, name='Dropout2'))

model.add(Dense(32, activation='relu', name='Dense3'))
model.add(BatchNormalization(name='BN3'))
model.add(Dropout(0.3, name='Dropout3'))

model.add(Dense(16, activation='relu', name='Dense4'))
model.add(BatchNormalization(name='BN4'))
model.add(Dropout(0.3, name='Dropout4'))

model.add(Dense(1, activation='sigmoid', name='Output'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
history = model.fit(X_train, y_train, epochs=200, batch_size=10)
```

```python
y_pred_prob = model.predict(X_test)
y_pred = (y_pred_prob > 0.5).astype(int)

accuracy = accuracy_score(y_test, y_pred)
print(f"test acc: {accuracy:.4f}")
```
