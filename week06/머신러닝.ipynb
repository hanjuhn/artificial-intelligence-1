### 전처리

```python
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold
from matplotlib import pyplot
from sklearn.preprocessing import StandardScaler

import pandas as pd
import matplotlib.pyplot as plt
```

```python
!git clone https://github.com/taehojo/data.git
```

```python
df = pd.read_csv('./data/pima-indians-diabetes3.csv')
df.head(5)
```

```python
X = df.iloc[:, 0:8]
y = df.iloc[:, 8]
```

```python
ss = StandardScaler()

scaled_X = pd.DataFrame(ss.fit_transform(X), columns=X.columns)
scaled_X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state= 0)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

### Decision Tree

```python
classifier = DecisionTreeClassifier()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
# 정확도(Accuracy)를 계산하는 아주 기본적인 방법
import numpy as np
np.sum(y_pred == np.array(y_test)) / len(y_pred)
```

```python
y_test
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
classifier = RandomForestClassifier(n_estimators= 50)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
np.sum(y_pred == np.array(y_test)) / len(y_pred)
```

### K-Nearest Neighbor

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
classifier = KNeighborsClassifier(n_neighbors= 5)
classifier.fit(X_train, y_train)
```

```python
np.sum(y_pred == np.array(y_test)) / len(y_pred)
```

### Ada Boost

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
classifier = AdaBoostClassifier()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
np.sum(y_pred == np.array(y_test)) / len(y_pred)
```

### Support Vector Machine

```python
from sklearn.svm import SVC
```

```python
classifier = SVC(kernel='linear')
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
np.sum(y_pred == np.array(y_test)) / len(y_pred)
```

```python
classifier = SVC(kernel='rbf')
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
np.sum(y_pred == np.array(y_test)) / len(y_pred)
```

### Voting (단일 모델은 아니고 위에서 배운 모델을 추가로 앙상블하는 것)

```python
from sklearn.ensemble import VotingClassifier
```

```python
clf1 = AdaBoostClassifier()
clf2 = RandomForestClassifier()
clf3 = SVC(kernel='linear')
classifier = VotingClassifier(estimators=[('lr', clf1), ('rf', clf2), ('gnb', clf3)])
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
np.sum(y_pred == np.array(y_test)) / len(y_pred)
```
