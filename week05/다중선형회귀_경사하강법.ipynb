### 다중 선형 회귀(Multiple Linear Regression)를 경사하강법(Gradient Descent)으로 학습

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x1 = np.array([2, 4, 6, 8])
x2 = np.array([0, 4, 2, 3])
```

```python
y = np.array([81, 93, 91, 97])
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.scatter3D(x1, x2, y)
plt.show()
```

```python
a1 = 0
a2 = 0
b = 0
```

```python
lr = 0.01
```

```python
epochs = 2001
```

```python
n = len(x1)
```

```python
for i in range(epochs):
  # 현재 가중치로 예측한 값 y_pred를 계산
  y_pred = a1 * x1 + a2 * x2 + b
  # error는 실제값과 예측값의 차이
  error = y - y_pred

  # 평균제곱오차(MSE)의 편미분을 통해 각 변수의 변화량 계산
  a1_diff = (2/n) * sum(-x1 *(error))
  a2_diff = (2/n) * sum(-x2 *(error))
  b_diff = (2/n) * sum(-(error))

  # lr: 학습률(learning rate)
  # 경사하강법에 따라 가중치를 오차가 줄어드는 방향으로 조금씩 조정
  a1 = a1 - lr * a1_diff
  a2 = a2 - lr * a2_diff
  b = b - lr * b_diff

  # 100회마다 학습 중간 결과 출력
  if i % 100 == 0:
    print("epoch: %.f, 기울기1 = %.04f, 기울기2 = %.04f, 절편 %.04f" % (i, a1, a2, b))
```

```python
# 선형 회귀(linear regression) 모델의 예측식 (2차원 입력 데이터를 받아서 선형 결합 후 예측값을 출력하는 전형적인 선형 회귀 구조)
y_pred = a1 * x1 + a2 * x2 + b
```

```python
# x1이 0, x2가 10일 때의 선형 회귀 예측값을 계산
a1 * 0 + a2 * 10 + b
```
