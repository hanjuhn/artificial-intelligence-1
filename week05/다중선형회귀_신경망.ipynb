### 다중 선형 회귀(Multiple Linear Regression)를 신경망으로 학습

```python
import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x = np.array([[2,0], [4,4], [6,2], [8,3]])
y = np.array([81, 93, 91, 97])
```

```python
model = Sequential()
```

```python
# 1은 출력 뉴런이 1개인 완전연결층
# input_dim이 2라는 것은 입력 데이터 차원이 2라는 뜻
# activation이 linear라는 것은 활성화 함수 없이 선형 연산만 수행
model.add(Dense(1, input_dim=2, activation = 'linear'))
```

```python
# compile 함수는 최적화 방법(옵티마이저)과 손실 함수(loss function)를 지정
model.compile(optimizer='sgd', loss='mse')
```

```python
# fit 함수는 모델을 실제로 학습(training)
model.fit(x, y, epochs=1000)
```

```python
# 모델에 대한 예측값 출력
model.predict(np.array([[0,7]]))
```
