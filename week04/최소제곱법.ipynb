### 최소제곱법으로 회귀 모델 만들기

```python
import numpy as np
```

```python
x = np.array([2,4,6,8])
```

```python
y = np.array([81,93,91,97])
```

```python
# 평균 구하기
mx = np.mean(x)
my = np.mean(y)
```

```python
divisor = sum([(mx-i)**2 for i in x])
```

```python
def top(x,mx,y,my):
  d = 0
  for i in range(len(x)):
    d += (x[i]-mx) * (y[i]-my)
  return d
```

```python
dividend = top(x,mx,y,my)
```

```python
a = dividend / divisor
```

```python
a
```

```python
b = my - (mx * a)
```

```python
a*7+b
```

```python
fake_a = 2.5
```

```python
fake_b = 75
```

```python
def predict_fake(x):
  return fake_a*x + fake_b
```

```python
def predict(x):
  return a*x+b
```

```python
predict_fake(1)
```

```python
predict(1)
```

```python
predict_result = []
```

```python
for i in range(len(x)):
  predict_result.append(predict(x[i]))
  print("공부시간=%.f, 실제점수=%.f, 예측점수=%.f" % (x[i], y[i], predict_fake(x[i])))
```

```python
n = len(x)
```

```python
def mse(y, y_pred):

  return (1/n) * sum((y-y_pred)**2)
```

```python
mse(y, predict_result)
```
